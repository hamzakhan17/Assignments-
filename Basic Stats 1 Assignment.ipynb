# Here’s what each statistic tells us:

1 count: The number of non-null observations in the column. This tells you how many data points are available for each statistic.

2 mean: The average value of the data in the column. It’s calculated by summing all the values and dividing by the count of values. It gives a measure of the central tendency.

3 std : (Standard Deviation): This measures the amount of variation or dispersion in the data. A higher standard deviation means the data points are spread out more from the mean, while a lower standard deviation means they are closer to the mean.

4 min: The minimum value in the column. This shows the smallest observation in the dataset.

- For example ( Volume : count 450.00, mean 5.06667, std 4.231602, min 1.00000 ) and its vice versa for every column and so on.

## Standardization (z-score normalization) - 
Standardization, also known as z-score normalization, is a method to transform the data to have a mean of zero and a standard deviation of one. This is particularly useful in analytical models where the features should be on a similar scale.

The formula for standardization is:
z= x- 𝜇/𝜎
 
where:

𝑥 - x is the original value.
𝜇 - μ is the mean of the column.
𝜎 - σ is the standard deviation of the column.
𝑧 - z is the standardized value.

## Steps to Standardize Numerical Columns
1. Calculate the mean (𝜇) and standard deviation (σ) for each numerical column.
2. Apply the standardization formula to each value in the numerical columns.
3. Compare the data distributions before and after standardization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'file:///C:\Users\dell\AppData\Local\Temp\7926b913-9815-4677-aa96-19f4980d87b2_Basic%20stats%20-%201.zip.7b2\Basic%20stats%20-%201\sales_data_with_discounts.csv')


# Identify numerical columns
numerical_columns = data.select_dtypes(include='number').columns.tolist()

# Calculate mean and standard deviation
means = data[numerical_columns].mean()
std_devs = data[numerical_columns].std()

# Standardize the numerical columns
standardized_data = (data[numerical_columns] - means) / std_devs

# Plot histograms before and after standardization
fig, axes = plt.subplots(2, len(numerical_columns), figsize=(15, 10))

for i, column in enumerate(numerical_columns):
    # Original data histogram
    axes[0, i].hist(data[column], bins=30, alpha=0.7, color='blue')
    axes[0, i].set_title(f'Original {column}')
    axes[0, i].set_xlabel(column)
    axes[0, i].set_ylabel('Frequency')
    
    # Standardized data histogram
    axes[1, i].hist(standardized_data[column], bins=30, alpha=0.7, color='green')
    axes[1, i].set_title(f'Standardized {column}')
    axes[1, i].set_xlabel(column)
    axes[1, i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

- When standardizing data, the shape of the distribution does not change; only the scale of the data is altered. To better observe the changes in the data due to standardization, you can compare the summary statistics (mean and standard deviation) before and after standardization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'file:///C:\Users\dell\AppData\Local\Temp\7926b913-9815-4677-aa96-19f4980d87b2_Basic%20stats%20-%201.zip.7b2\Basic%20stats%20-%201\sales_data_with_discounts.csv')

# Identify numerical columns
numerical_columns = data.select_dtypes(include='number').columns.tolist()

# Calculate mean and standard deviation for numerical columns
means = data[numerical_columns].mean()
std_devs = data[numerical_columns].std()

# Standardize the numerical columns using the formula: z = (x - mu) / sigma
standardized_data = (data[numerical_columns] - means) / std_devs

# Print summary statistics before and after standardization
print("Summary Statistics Before Standardization:")
print(data[numerical_columns].describe())

print("\nSummary Statistics After Standardization:")
print(standardized_data.describe())

# Plot boxplots before and after standardization
fig, axes = plt.subplots(2, len(numerical_columns), figsize=(15, 10))

for i, column in enumerate(numerical_columns):
    # Original data boxplot
    axes[0, i].boxplot(data[column])
    axes[0, i].set_title(f'Original {column}')
    axes[0, i].set_xlabel(column)
    
    # Standardized data boxplot
    axes[1, i].boxplot(standardized_data[column])
    axes[1, i].set_title(f'Standardized {column}')
    axes[1, i].set_xlabel(column)

plt.tight_layout()
plt.show()
```

# Task4 - Conversion of Categorical Data into Dummy Variables

●	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
- Steps:
	- Discuss the need for converting categorical data into dummy variables (one-hot encoding).
	- Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
	- Display a portion of the transformed dataset.

```python
sales.columns
```

```python
# One-Hot Encoding - it is used to convert obj into int.
#                   - one-hot encoding creates binary columns for each category.
#                     allowing the model to interpret categorical data correctly without assuming any ordinal relationship.
```

```python
#1. Identify Categorical Columns

# Selecting categorical columns and storing them in cc
categorical_columns = sales.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_columns)
```

```python
sales.Date
```

```python
#Since 'Date' column is in the format 'DD-MM-YYYY' and is currently a string, we'll need to convert it to a datetime format.
```

```python
# Converting 'Date' to datetime format
sales['Date'] = pd.to_datetime(sales['Date'], format='%d-%m-%Y')   

# storing them seprately as Year=Year, Monthn=Month, Day=Day and weekadys as Monday=0, Tuesday=1 n so on...
sales['Year'] = sales['Date'].dt.year
sales['Month'] = sales['Date'].dt.month
sales['Day'] = sales['Date'].dt.day
sales['Weekday'] = sales['Date'].dt.weekday  # Monday=0, Sunday=6

# Drop the original 'Date' column
sales = sales.drop(columns=['Date'])
```

```python
sales.columns
```

```python
# Selecting categorical columns and storing them in cc
new_categorical_columns = sales.select_dtypes(include=['object']).columns
print("Categorical columns:", new_categorical_columns)
```

```python
# Applyin one-hot encoding to the specified categorical columns
sales = pd.get_dummies(sales, columns=new_categorical_columns)


# Display the first few rows of the transformed dataset
sales.head()
```
